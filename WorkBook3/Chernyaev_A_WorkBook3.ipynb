```python
import numpy as np
import matplotlib.pyplot as plt


def sqr_euclidean_distance(p1, p2):
    return sum((x-y)**2 for x, y in zip(p1, p2))

def weighted_euclidean_distance(p1, p2, w):
    return sum((x-y)**2*s for x, y, s in zip(p1, p2, w))**0.5

def manhattan_distance(p1, p2):
    return sum(abs(x-y) for x, y in zip(p1, p2))

def chebushev_distance(p1, p2):
    return max(abs(x-y) for x, y in zip(p1, p2))

p1 = np.array([0, 0, 0])
p2 = np.array([3, 3, 3])
p3 = np.array([0, 5, 5])
p4 = np.array([7, 6, 3])
w = np.array([0, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(p1[0], p1[1], p1[2])
ax.scatter(p2[0], p2[1], p2[2])
ax.scatter(p3[0], p3[1], p3[2])
ax.scatter(p4[0], p4[1], p4[2])
plt.show()

print("sqr_euclidean_distance(p1, p2) = ",
      sqr_euclidean_distance(p1, p2))
print("weighted_euclidean_distance(p2, p3, w) = ",
      weighted_euclidean_distance(p2, p3, w))
print("manhattan_distance(p3, p4) = ",
      manhattan_distance(p3, p4))
print("chebushev_distance(p4, p1) = ",
      chebushev_distance(p4, p1))
```

# 2.3.1

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns


iris = sns.load_dataset('iris')

X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size = 0.15
)
k = 1
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width',
                y='petal_length',
                data=iris,
                hue='species',
                s=70)
plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend(loc=2)
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='red', s=150)
        
print(f'accuracy: {accuracy_score(y_test, y_pred) :.3}')
```
