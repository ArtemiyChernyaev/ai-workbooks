```python
import numpy as np

rows = 8
columns = 8

mat = np.zeros(rows*columns).reshape(rows, columns)

for n in range(rows):
    for m in range(n%2, columns, 2):
        mat[n, m] = 1
        
print(mat)
```

# 1.3.4

```python
import numpy as np

rows = 8
columns = 8

mat = np.zeros(rows*columns).reshape(rows, columns)

for i in range(columns):
    mat[0, i] = 1
for i in range(columns):
    mat[rows-1, i] = 1
for i in range(rows):
    mat[i, 0] = 1
for i in range(rows):
    mat[i, columns-1] = 1
        
print(mat)
```

# 3.3.2

```python
import pandas as pd
import numpy as np
from math import sqrt

url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"

tab = pd.read_csv(url)

xmin = tab['sepal_length_cm'].min()
xmax = tab['sepal_length_cm'].max()

std = np.std(tab['sepal_width_cm'])
mean = np.mean(tab['sepal_width_cm'])

def norm_minmax(x):
    return (x-xmin)/(xmax-xmin)

def z_scaling(x):
    return (x - mean)/std

tab['sepal_length_cm'] = tab['sepal_length_cm'].apply(norm_minmax)
tab['sepal_width_cm'] = tab['sepal_width_cm'].apply(z_scaling)

tab
```
