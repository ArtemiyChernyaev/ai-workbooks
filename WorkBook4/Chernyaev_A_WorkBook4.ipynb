```python
from numpy import *
from numpy.random import *

delta = 1
x = linspace(-5, 10, 16)
y = x**2 + delta*(rand(16)-0.5)
x += delta*(rand(16)-0.5)

m = vstack((x**2, x, ones(16))).T
s = np.linalg.lstsq(m, y , rcond=None)[0]

x_prec = linspace(-5, 10, 16)
plt.plot(x, y, 'D')
plt.plot(x_prec, s[0]*x_prec**2 + s[1]*(x_prec) + s[2])
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit

beta = (1, 2, 0.2)

def f(x, b0, b1, b2):
    return b0 + b1*np.exp(-b2*x**2)

xdata = linspace(0, 5, 100)
y = f(xdata, *beta)
ydata = y + 0.05 * np.random.randn(len(xdata))

beta_opt, beta_cov = curve_fit(f, xdata, ydata)

lin_dev = sum(beta_cov[0])
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r')
ax.plot(xdata, f(xdata, *beta_opt), 'b')
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize = 18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize = 18)

plt.grid()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

file = pd.read_csv("https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv")

X = file.iloc[:, :-1].values
y = file.iloc[:, 1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

plt.scatter(X, y)
plt.plot(X_test, y_pred, color='orange')
plt.show()
pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

file = pd.read_csv("https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv")

plt.scatter(file['Average_income'], file['Petrol_Consumption'])

X = file.iloc[:, 1].values.reshape(-1, 1)
y = file.iloc[:, 4].values.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

print("k =", regressor.coef_[0][0])
print("b =", regressor.intercept_[0])
plt.scatter(X, y)
plt.plot(X_test, y_pred)
plt.show()
pd.DataFrame({'y_test': np.concatenate(y_test), 'y_pred': np.concatenate(y_pred)})
```
